### Let's Clean up data

```python
def textClean(string):
  try:    
    string = re.sub(r'^https?:\/\/<>.*[\r\n]*', '', string, flags=re.MULTILINE)
    string = re.sub(r"[^A-Za-z]", " ", string)         
    words = string.strip().lower().split()    
    words = [w for w in words if len(w)>=1]
    return " ".join(words)	
  except:
    return ""
```

```python
# Use apply function to clean review data and store it to same data frame
df['cleanArticle'] = df['news_article'].apply(textClean) 
```

### Convert Review to a Word List

```python
#List to hold all words in each review
documents = []
```

```python
#Iterate over each review
for doc in df['cleanArticle']:
    documents.append(doc.split(' '))
```

```python
len(documents[23])
```

```python
#documents[23]
```

### Train the Model

```python
model = gensim.models.Word2Vec(documents, #Word corpus
                               min_count=10, #Ignore all words with total frequency lower than 10                           
                               size=50,  #Embedding size
                               window=5, #Maximum Distance between current and predicted word
                               iter=10   #No of iterations over the text corpus
                              )  
```
